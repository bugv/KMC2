```python
d_raw = np.genfromtxt("sampling.txt")

d = {}
qtys = ["total_00","r2s_00","crosses_00","total_11","r2s_11","crosses_11","theprod_01"]
for i,qty in enumerate(qtys) :
    key = qty + "_avg"
    d[key] = d_raw[2*i,:]
    key = qty + "_std"
    d[key] = d_raw[2*i+1,:]

d
```

```python
#analytical expression
Gamma = 364990214.518247042 # Hz
a = 4 / 2**0.5 # AA
x_V = 0.002 
x_B = 0.5
x_A = 1-x_B
z = 12
d = 3 
rho = z / 2 / d
f = 0.7815

nat = 500

L_00 = x_V * x_A * rho * a**2 * Gamma * (1 - x_B / (1-x_V) *  (1-f))
predict_00 = L_00 * np.arange(1000) * 6 * 500 * 2.28023874e-04 / 1000

L_11 = x_V * x_B * rho * a**2 * Gamma * (1 - x_A / (1-x_V) *  (1-f))
L_01 = x_V * x_A * x_B / (1- x_V) * rho * a**2 * Gamma * (1-f)
predict_01 = L_01 * np.arange(1000) * 6 * 500 * 2.28023874e-04 / 1000
predict_11 = L_11 * np.arange(1000) * 6 * 500 * 2.28023874e-04 / 1000
```

```python
predict_r2 = np.arange(1000) * 6 * 500 * 2.28023874e-04 / 1000 * 2 * f * x_V * (rho * a**2 * Gamma) / 4
predict_r2
```

```python
plt.plot(d["r2s_00_avg"],label="mean R^2 terms (sum over all sites)")
plt.plot(d["crosses_00_avg"], label="mean cross terms")
plt.plot(d["total_00_avg"], label="mean total")

plt.fill_between(range(len(d["r2s_00_avg"])), 
                 d["r2s_00_avg"] - d["r2s_00_std"], 
                 d["r2s_00_avg"] + d["r2s_00_std"], 
                 alpha=0.1)

plt.fill_between(range(len(d["crosses_00_avg"])), 
                 d["crosses_00_avg"] - d["crosses_00_std"], 
                 d["crosses_00_avg"] + d["crosses_00_std"], 
                 alpha=0.3)

plt.plot(predict_00, label="predicted")
plt.plot(predict_01, label="predicted 01 cross term")
plt.plot(d["theprod_01_avg"].mean(axis=0), label="KMC 01 mean cross term")
plt.title(f"L_00 terms over {d['r2s_00_avg'].shape[0]} runs")

plt.plot(predict_r2,label="hi")

plt.legend()
plt.ylabel("AA^2")
plt.xlabel("propto time")
plt.grid()
plt.show()
```
